```python
links = []
numberoflinks = 0
for link in soup.findAll( 'a', attrs={'href': re.compile("^https:")}):
    links.append(link.get('href'))
    numberoflinks = numberoflinks + 1
#print(numberoflinks)
#print(links)
```

```python
#remove_list = ['TandC','doi.org', 'myaccount']
links = [e for e in links if "TandC" not in e and "doi.org" not in e and "myaccount" not in e]
print(links)
```
